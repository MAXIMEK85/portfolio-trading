```python
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import pandas as pd

#Actions choisies + periode
equities = ["MC.PA","MMM","PYPL","NKE","BABA","JD"]
period = 14
for ticket in equities:

  #Recuperer les valeurs + pouvoir les utiliser (close)
  df= yf.download(ticket, start="2025-06-01")
  print(df.columns)
  close = df[("Close",ticket)] if isinstance(df.columns, pd.MultiIndex) else df["Close"]
  close = close.dropna()

  # CALCUL RSI
  delta = close.diff()
  gain = delta.clip(lower=0)
  loss = -delta.clip(upper=0)
  avg_gain = gain.ewm(alpha=1/period,adjust=False).mean()
  avg_loss = loss.ewm(alpha=1/period,adjust=False).mean()
  rs = (avg_gain / avg_loss)
  rsi = 100 - (100 / (1+ rs))

  # RSI MIN ; RSI MAX
  print("RSI min", rsi.min())
  print("RSI max", rsi.max())

  #CALCUL SMA EMA
  SMA = df["Close"].rolling(window=200).mean()
  EMA = close.ewm(span=20, adjust=False).mean()

  #CALCUL MACD
  ema12 = close.ewm(span=12,adjust=False).mean()
  ema26 = close.ewm(span=26,adjust=False).mean()
  MACD = ema12 - ema26
  MACD_SIGNAL = MACD.ewm(span=9,adjust=False).mean()
  IMACD = MACD - MACD_SIGNAL
  peaks,_=find_peaks(close)

  # Affichage RSI
  plt.plot(rsi, label="RSI")
  plt.title(f"RSI: {ticket}")
  plt.xlabel("time")
  plt.ylabel("%")
  plt.axhline(50, color="black", label="Mediane")
  plt.axhline(25, color="purple", label="25%")
  plt.axhline(75, color="yellow", label="75%")
  plt.legend()
  plt.figure()

  #affichage MACD
  plt.plot(MACD, label="MACD")
  plt.plot(MACD_SIGNAL, label="MACD_SIGNAL")
  plt.legend()
  plt.figure()

  # Affichage
  plt.hist(IMACD, bins=200)
  summary = IMACD.describe()
  summary["Ticket"]=ticket
  print(summary)
  plt.figure()

  # Affichage SMA EMA
  plt.plot(SMA,label ="SMA")
  plt.plot(EMA,label ="EMA",color="red")
  plt.plot(close.index, close, label="Close Price")
  plt.plot(close.index[peaks], close.iloc[peaks], "yo", label="Peaks")
  plt.title(ticket)
  plt.xlabel("time")
  plt.ylabel("price")
  plt.legend()
  plt.figure()

  # Histogramme Volume
  plt.hist(df["Volume"], bins=200)
  plt.title("Volume")
  plt.xlabel("nombre de jours")
  plt.ylabel("Taille du volume")
  plt.show()

  #Description du volume
  print(df["Volume"].describe())

  #
  high_volume = df[df["Volume"] > df["Volume"].mean()]
  print(high_volume["Close"])
  t = df["Volume"].pct_change()
  print(t)

  #buy = rsi[rsi <35]
  #for date, value in buy.items():
  #  print(f"{ticket}BUY on {date.date()}")

  #sell = rsi[rsi >70]
  #for date, value in sell.items():
  #  print(f"{ticket}BUY on {date.date()}-RSI = {value:.1f}")


  #liste pour stocker les PnL
  #PARAMETRES STRATEGIES


  trade_pnls = []
  in_long = False
  in_short = False
  buy_price= None
  target_price = None
  target_price_short = None

  buy_level = 35
  take_profit = 0.05
  sell_level = 70



  #BACKTEST


  for date, value in rsi.items():

    price = float(df.loc[date,"Close"])

    # BUY ACHAT

    if value < buy_level and not in_position:
        buy_price = price
        target_price = buy_price * (1+ take_profit)
        in_long =True
        print(f" BUY on {date} @ {buy_price:.2f} (RSI = {value:.1f})")

    # SELL

    elif in_long and price >= target_price:
        trade_pnl = price - buy_price
        trade_pnls.append(trade_pnl)
        in_long = False
        buy_price = None
        target_price= None
        print(f" SELL on {date.date()} @ {price:.2f} | Pnl ={trade_pnl:.2f}")


      # SHORT

    if value > sell_level and not in_long and not in_short:
        short_price = price
        target_price_short = short_price * (1 - take_profit)
        in_short = True
        print(f" SHORT ")

    elif in_short and price <= target_price_short:
        trade_pnl = short_price - price
        trade_pnls.append(trade_pnl)
        in_short =False
        short_price = None
        target_price_short = None
        print(f" SELL SHORT")



  #RESULTATS
  if len(trade_pnls) > 0:
      total_pnl = sum(trade_pnls)

      print("\n---- RESULTS -----")
      print(f"Trades excecuted: {len(trade_pnls)}")
      print(f"Total PnL: {total_pnl:.2f}")
      print(f"Average PnL per trade: {total_pnl/len(trades):.2f}" if trades else "No trades")
  else:
    print("\n----RESULTS-----")
    print("NO TRADES")
```
