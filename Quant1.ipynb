```python
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import pandas as pd

equities = ["MC.PA","MMM","PYPL","NKE","BABA","JD"]
period = 14
for ticket in equities:

  df= yf.download(ticket, start="2020-01-01")
  print(df.columns)
  close = df[("Close",ticket)] if isinstance(df.columns, pd.MultiIndex) else df["Close"]
  close = close.dropna()

  #RSI
  delta = close.diff()
  gain = delta.clip(lower=0)
  loss = -delta.clip(upper=0)
  avg_gain = gain.ewm(alpha=1/period,adjust=False).mean()
  avg_loss = loss.ewm(alpha=1/period,adjust=False).mean()
  rs = (avg_gain / avg_loss)
  rsi = 100 - (100 / (1+ rs))

  SMA = df["Close"].rolling(window=200).mean()

  peaks,_=find_peaks(close)

  plt.plot(rsi, label="RSI")
  plt.title(ticket)
  plt.xlabel("time")
  plt.ylabel("%")
  plt.axhline(50, color="black", label="Mediane")
  plt.axhline(25, color="purple", label="25%")
  plt.axhline(75, color="yellow", label="75%")
  plt.legend()
  plt.figure()
  plt.plot(SMA,label ="SMA")
  plt.plot(close.index, close, label="Close Price")
  plt.plot(close.index[peaks], close.iloc[peaks], "yo", label="Peaks")

  plt.title(ticket)
  plt.xlabel("time")
  plt.ylabel("price")
  plt.legend()
  plt.figure()

  plt.hist(df["Volume"], bins=200)
  plt.title("Volume")
  plt.xlabel("nombre de jours")
  plt.ylabel("Taille du volume")
  plt.show()


  print(df["Volume"].describe())

  high_volume = df[df["Volume"] > df["Volume"].mean()]
  print(high_volume["Close"])

  t = df["Volume"].pct_change()
  print(t)
```
